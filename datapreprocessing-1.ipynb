importing necessary libraries

```python
import pandas as pd
import numpy as np 
```

creating dataframe

```python
data = {
    'Student_ID': ['220967', '220968', '220969', '220970', '220971', '220972', '220973', np.nan, '220975', '220976'],
    'Name': ['Aleeza', 'Taha', 'amina', 'awais', 'aqsa', 'eraj', 'aima', 'Haris', 'hanniya', 'ehtisham'],
    'Gender': ['F', 'M', np.nan, 'M', np.nan, 'M', 'F', 'F', 'F', 'M'],
    'Age': [20, 21, np.nan, 22, 20, 20, 21, 19, np.nan, 22],
    'mids': [85, 90, np.nan, 78, 92, 88, 85, 91, 80, 87],
    'internals': [88, 84, 91, 85, 80, np.nan, 89, 92, 81, 90],
    'project': [89, 85, 91, np.nan, np.nan, 86, 90, 88, 84, 91],
    'finals': [90, 80, 85, 88, 91, 87, np.nan, 86, 89, 90],
    'Passed': ['Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes'] 
}

df = pd.DataFrame(data)
df
```

1- filling null values using mean,median,mode imputation

```python
df['imputed_meanmids'] = df['mids'].fillna(df['mids'].mean())
df
```

```python
df['imputed_medianproject'] = df['project'].fillna(df['project'].median())
df
```

```python
#for numnerical data
df['imputed_modefinals'] = df['finals'].fillna(df['finals'].mode())
#for categorical data 
df['imputed_mode gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df
```

2- Removing rows

```python
removed_rows_df = df.dropna()  
removed_rows_df
```

3- Interpolation

```python
# numerical-Interpolation
df['interpolated_age'] = df['Age'].interpolate(method='linear')
df
```

4-Forward and backward fills

```python
df['project_forwarded'] = df['project'].fillna(method='ffill')
df['project_backwarded'] = df['project'].fillna(method='bfill')
df
```
